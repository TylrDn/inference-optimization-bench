# Quantization Comparison
Compare fp16 vs GPTQ 4-bit vs AWQ 4-bit vs GGUF Q4_K_M across latency, throughput, and VRAM.

```python
import pandas as pd, glob, plotly.express as px
dfs = [pd.read_csv(f) for f in glob.glob('../results/*.csv')]
df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()
print(df.describe() if not df.empty else 'No results yet — run bench/run_bench.py first')
```

```python
# P95 latency bar chart
if not df.empty:
    fig = px.bar(df.groupby('_file')['total_latency_ms'].quantile(0.95).reset_index(),
                 x='_file', y='total_latency_ms', title='P95 Latency by Run')
    fig.show()
```
